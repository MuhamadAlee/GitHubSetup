```python
# importing Libraries

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

import matplotlib as mpl
mpl.rcParams["figure.facecolor"]="white"
```

```python
# loading the dataframe 
tips_df = sns.load_dataset('tips')

tips_df.head()
```

## Count Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.countplot(x="smoker", hue="sex", data=tips_df)



# labelling the plot
plt.xlabel("smokers counts")
plt.ylabel("Male and Female Smokers")

plt.title("Smokers counts")
```

## Bar plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.barplot(x="day", y="total_bill", hue="sex", data=tips_df)



# labelling the plot
plt.xlabel("Days")
plt.ylabel("total bill")

plt.title("Smokers in Week Days")
```

## Box Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.boxplot(y="total_bill", x="sex", hue="time", data=tips_df, showmeans=True)

# labelling the plot
plt.xlabel("sex")
plt.ylabel("total bill")

plt.title("Smokers in different times")
```

## Violin Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.violinplot(y="total_bill", x="sex", hue="time", data=tips_df, showmeans=True)

# labelling the plot
plt.xlabel("sex")
plt.ylabel("total bill")

plt.title("Smokers in different times")
```

## CAT Plot

```python
#it allows to try different plots using this one plot eg bar, violin and so on

# setting the figure size
plt.figure(figsize=(15,5))

sns.catplot( x="day", hue="smoker", col="day", kind="count", data=tips_df,
            col_wrap=2, facet_kws=dict(sharex=False, sharey=False))

# labelling the plot
plt.xlabel("Day")
plt.ylabel("total bill")

plt.title("Smokers in different times")
```

## Histograms

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.histplot(x="total_bill", hue="sex", data=tips_df, kde=True, multiple="stack", bins=[5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
plt.xticks([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])

# labelling the plot
plt.xlabel("stacked ")
plt.ylabel("Frequency")

plt.title("Smokers in different times")
```

## Dis Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.displot(x="total_bill", hue="sex", col="day", data=tips_df, kde=True,
            col_wrap=2, facet_kws=dict(sharey=False, sharex=False))

# labelling the plot
plt.xlabel("stacked ")
plt.ylabel("Frequency")

plt.title("Smokers in different times")
```

## Pair Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.pairplot(tips_df, hue="sex", diag_kind="kde")  

# labelling the plot
plt.title("Pair plots")
```

## Scatter Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.scatterplot(x="total_bill", y="tip", hue="sex", style="day", data=tips_df)

# labelling the plot
plt.title("Scatter plots")
```

## Regression Plot


```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.regplot(x="tip", y="total_bill", data=tips_df, ci=False)

# labelling the plot
plt.title("Regression plots")
```

## Line Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.lineplot(x="day", y="total_bill", hue="sex", data=tips_df, ci=False, marker="o")

plt.grid()

# labelling the plot
plt.title("Line plots")
```

## Rel Plot

```python
# setting the figure size
plt.figure(figsize=(15,5))

sns.relplot(x="total_bill", y= "tip", hue="sex", style='day', data=tips_df)

plt.grid()

# labelling the plot
plt.title("Relational plots")
```

## Heatmap

```python
# setting the figure size
plt.figure(figsize=(15,5))

# visualize the correlation matrix
sns.heatmap(data=tips_df.corr(),annot=True)
```

## Facet Grid

```python
sns.set_palette("afmhot")

matrix = sns.FacetGrid(data=tips_df, row="time", col="day")
matrix.map(sns.countplot, x="smoker", hue="sex", data=tips_df).add_legend()

```
