```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib as mpl
mpl.rcParams['figure.facecolor'] = 'white'
```

```python
# setting up the dataframe

data={
    "Country" : ["United States", "Pakistan", "India", "United Kingdom"],
    "Gold_Medals" : [45, 3, 20, 40],
    "Silver_Medals" : [50, 20, 58, 50],
    "Browns_Medals" : [90, 40, 40, 69]
}

df=pd.DataFrame(data)
```

## Bar Chart

```python
#styling the plot
plt.style.use('tableau-colorblind10')

#setting the size of the plot
plt.figure(figsize=(15,5))

# making bars for single column or series
plt.bar(df["Country"], df["Gold_Medals"], width=0.5, color="green")

# labeling x and y axis
plt.xlabel("Countries")
plt.ylabel("Gold Medals")

# titling the plot
plt.title("Countries with gold medals")
```

```python
#styling the plot
plt.style.use('seaborn-notebook')

#setting the size of the plot
plt.figure(figsize=(15,5))

# making bars for multiple columns
x_axis=np.arange(len(df["Country"]))

plt.bar(x_axis-0.2, df["Gold_Medals"], width=0.3, color="green")
plt.bar(x_axis, df["Silver_Medals"], width=0.3, color="blue")
plt.bar(x_axis+0.2, df["Browns_Medals"], width=0.3, color="orange")


# replacing the numbers with categories
plt.xticks(x_axis, df["Country"])

# labeling x and y axis
plt.xlabel("Countries")
plt.ylabel("Medals")

# titling the plot
plt.title("Countries with gold medals")

# legends to determine which color belongs to which

plt.legend(["Gold Medal", "Silver Medal", "Browns Medal"])


```

## Pie Chart

```python
#styling the plot
plt.style.use('_classic_test_patch')

# setting the figure size 
plt.figure(figsize=(15, 5))

plt.pie(df["Gold_Medals"],labels = df["Country"], autopct="%1.2f%%", shadow=True)

# titling the plot
plt.title("Countries with gold medals")
```

## Scatter chart

```python
#styling the plot
plt.style.use('seaborn-whitegrid')

# setting up the figure size for the plot
plt.figure(figsize=(15,5))

# scatter plot for all the medals along with the country
plt.scatter(df["Country"], df["Gold_Medals"], marker="^", s=200, color="red")
plt.scatter(df["Country"], df["Silver_Medals"], marker="x", s=200, color="blue")
plt.scatter(df["Country"], df["Browns_Medals"], marker="v", s=200, color="orange")

#labelling the plot
plt.xlabel("Countries")
plt.ylabel("Number of medals")

# title of the plot
plt.title("Countries' medals distributions")


# setting up the legends
plt.legend(["Gold_Medals", "Silver_Medals", "Browns_Medal"])

```

## Box Plot

```python
# setting up the figure size
plt.figure(figsize=(15,5))

# box plot for medals
plt.boxplot(df[["Gold_Medals", "Silver_Medals", "Browns_Medals"]],
            showmeans=True, meanline=True, vert=True,
            labels=["Gold_labels", "Silver_Models", "Browns_Models"],
            )

# labeling the plot
plt.title("box plots for all medals")
```

## Line Plot

```python
#styling the plot
plt.style.use('seaborn-darkgrid')

# setting the plot size
plt.figure(figsize=(15,5))

# line plot
plt.plot(df["Country"], df[["Gold_Medals", "Silver_Medals", "Browns_Medals"]], 
         marker="o", markersize=15, linewidth=8, markerfacecolor="white")

# make grids
# plt.grid()

plt.title("medals attained by the countries")

# labeling
plt.xlabel("Countries")
plt.ylabel("Number of Medals")

#legends
plt.legend(["Gold_Medals", "Silver_Medals", "Browns_Medals"])

```
